## How many products are being purchased more than once?

## Do the products being purchased again have better sales than others?
The products purchased once have higher total sales (51125000 dollars) compared to products that were purchased more than once (45587000 dollars). However, products purchased more than once have higher average sales compared to products purchased once. Products in the category of Snacks, toys and supplements that were purchased once have higher total sales than those that were purchased more than once. This could be due to the large prices of these products. The marketing team may want to include these onetime purchase products in their list.

## What products are more likely to be purchased again for different types of pets?
The products that have been purchased more than once are more likely to be purchased again for different types of pets. For pet dogs and cats, products such equipment, food, medicine, toys and snacks are more likely to be purchased again. Equipment products, snacks and toys are more likely to be repurchased for fish pets. Pet birds have only three possible products that are more likely to be purchased again, these are equipment, toys and snack products. Based on this, equipment products, toy products and snack products should be added to the list of product subscription as they are more likely to be purchased again by customers.
